Model Integration With OPENAI,GEMNI,GOOGLE and GROQ

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

```

Groq model integration

```python
from langchain.chat_models import init_chat_model

# Groq ka sahi setup
model = init_chat_model(
    model="llama-3.1-8b-instant",  # <-- 'model_name' ki jagah sirf 'model' likhein
    model_provider="groq",           # <-- Provider batana zaroori hai
    temperature=0.6
)

response = model.invoke("Hello, how are you?")
print(response.content)
```

Google Gemni Model Integration

```python
from langchain.chat_models import init_chat_model

# Google Gemini ka updated setup
model = init_chat_model(
    model="gemini-2.5-flash",        # <-- 'gemini-1.5-flash' ki jagah latest 'gemini-2.5-flash' use karein
    model_provider="google_genai",
    temperature=0.6 
)

response = model.invoke("Hello, How are you?")
print(response.content)
```

chatGroq

```python
from langchain_groq import ChatGroq

model = ChatGroq(
    model_name = "llama-3.1-8b-instant",
    temperature=0.6
)



response = model.invoke("Hello , how are you ?")
response.content 
```

chatGemni

```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI

# 'model_name' ki jagah sirf 'model' likhein aur 'gemini' ki spelling sahi karein
model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",  # <-- Fixed parameter name and spelling!
    temperature=0.6
)

response = model.invoke("Why do parrots talk?")
print(response.content)
```

Streaming

```python
response = model.invoke("write me 200 words paragraph on artificial intelligence")
response.content
```

```python
for chunk in model.stream("Write a 200 paragraph on artificial intelligence"):

    print(chunk.text,end = "",flush = True)
```

```python
for chunk in model.stream("Why do parrots have colourful feathers ?"):
    print(chunk.text,end ="|",flush = True)


```

Batch :-If you have list of inputs and do ont want to create seprate pipeline for each then you can use the batch

```python
input = [
    "what is capital of france ?",
    "who is was the president of us in 2020 ?",
    "who is the president of Oman ?"

]



response = model.batch(input)

for i,res in enumerate(response):
    print(f"{i+1} {res.content}")
```

```python
model.batch(
    [ "why do parrots have colourful feather ? ",
     "How do airplanes fly ? ",
     "What is quantum computing ? "],



config = {
    "max_concurrency":5,
})
```
